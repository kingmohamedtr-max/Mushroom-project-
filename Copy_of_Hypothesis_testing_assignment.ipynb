```python
# Load the datasets
gdp_data = pd.read_excel('GDP per capita (5).xlsx')
whr_data = pd.read_excel('WHR2023_Data-1 (5).xlsx')
# Inspect the column names and the first few rows to understand their structure
print("GDP per capita (5).xlsx:", gdp_data.columns)
print("WHR2023_Data-1 (5):", whr_data.columns)
# Rename columns to ensure they match for merging
gdp_data = gdp_data.rename(columns={'Country': 'Country Name'}) # Match 'Country' to 'Country Name'
whr_data = whr_data.rename(columns={'Country name': 'Country Name'}) # Match 'Country name' to 'Country Name'
# Merge the datasets on 'Country Name' only (no 'Year' column in the WHR data)
merged_data = pd.merge(gdp_data, whr_data, on='Country Name', how='inner')
# Check the merged data
print(merged_data.head())

```

```python
# Plotting GDP per Capita
plt.hist(merged_data['GDP per capita'], bins=20, color='blue', edgecolor='black')
plt.title('Distribution of GDP per Capita')
plt.xlabel('GDP per Capita')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting Happiness
plt.hist(merged_data['Happiness Index'], bins=20, color='green', edgecolor='black')
plt.title('Distribution of Happiness Index')
plt.xlabel('Happiness Index')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot with trend line showing the correlation between GDP per capita and Happiness Index
plt.figure(figsize=(10, 6))
sns.regplot(x='GDP per capita', y='Happiness Index', data=merged_data, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
# Adding labels and title
plt.title('Correlation between GDP per Capita and Happiness Index')
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Index')
# Show plot
plt.show()
```

```python
#Calculate Variance and Standard Deviation
gdp_variance = np.var(merged_data['GDP per capita'])
gdp_stddev = np.std(merged_data['GDP per capita'])
happiness_variance = np.var(merged_data['Happiness Index'])
happiness_stddev = np.std(merged_data['Happiness Index'])
print("GDP per Capita Variance:", gdp_variance)
print("GDP per Capita Standard Deviation:", gdp_stddev)
print("Happiness Index Variance:", happiness_variance)
print("Happiness Index Standard Deviation:", happiness_stddev)
```

```python
# Calculate Z-scores for GDP per capita and Happiness Index
merged_data['GDP Z-score'] = (merged_data['GDP per capita'] - np.mean(merged_data['GDP per capita'])) / np.std(merged_data['GDP per capita'])
merged_data['Happiness Z-score'] = (merged_data['Happiness Index'] - np.mean(merged_data['Happiness Index'])) / np.std(merged_data['Happiness Index'])
print('GDP Z-score')
print(merged_data['GDP Z-score'])
print('Happiness Z-score')
print(merged_data['Happiness Z-score'])
```

```python
from scipy.stats import norm

# Example: Calculate the probability of a GDP per capita value being less than 30000
gdp_mean = np.mean(merged_data['GDP per capita'])
gdp_std = np.std(merged_data['GDP per capita'])
gdp_value = 30000
probability_below_30000 = norm.cdf(gdp_value, loc=gdp_mean, scale=gdp_std)
print(f"Probability of GDP per capita < {gdp_value}: {probability_below_30000}")


# Example: Calculate the probability of a Happiness Index value being between 6 and 7
happiness_mean = np.mean(merged_data['Happiness Index'])
happiness_std = np.std(merged_data['Happiness Index'])
happiness_lower = 6
happiness_upper = 7
probability_between_6_and_7 = norm.cdf(happiness_upper, loc=happiness_mean, scale=happiness_std) - norm.cdf(happiness_lower, loc=happiness_mean, scale=happiness_std)
print(f"Probability of Happiness Index between {happiness_lower} and {happiness_upper}: {probability_between_6_and_7}")
```

```python
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt


# Prepare the data for regression
X = merged_data['GDP per capita'] # Independent variable (GDP per capita)
y = merged_data['Happiness Index'] # Dependent variable (Happiness Index)
# Add a constant to the independent variable (for the intercept)
X = sm.add_constant(X)
# Fit the regression model
model = sm.OLS(y, X).fit()
# Print the regression summary
print(model.summary())
# Plotting the Regression Line on the scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=merged_data, x='GDP per capita', y='Happiness Index', color='blue', label='Data Points')
# Plot the regression line
plt.plot(merged_data['GDP per capita'], model.predict(X), color='red', label='Regression Line')
plt.title("Multiple Regression: GDP per Capita vs. Happiness Index")
plt.xlabel("GDP per Capita")
plt.ylabel("Happiness Index")
plt.legend()
```

```python
# hypothesis testing
import scipy.stats as stats
# Perform a two-sample t-test
t_statistic, p_value = stats.ttest_ind(merged_data['GDP per capita'], merged_data['Happiness Index'])
print("T-statistic:", t_statistic)
print("P-value:", p_value)
```

```python
# prompt: make an hypothesis testing

# Define the significance level (alpha)
alpha = 0.05

# Null hypothesis: There is no significant difference between the means of 'GDP per capita' and 'Happiness Index'.
# Alternative hypothesis: There is a significant difference between the means of 'GDP per capita' and 'Happiness Index'.

# Check the p-value against the significance level
if p_value < alpha:
    print("Reject the null hypothesis.")
    print("There is a statistically significant difference between the means of 'GDP per capita' and 'Happiness Index'.")
else:
    print("Fail to reject the null hypothesis.")
    print("There is no statistically significant difference between the means of 'GDP per capita' and 'Happiness Index'.")
```

```python
# Assuming 'merged_data' DataFrame is already created and populated as in your previous code.

import scipy.stats as stats

# Extract the relevant columns
gdp = merged_data['GDP per capita']
happiness = merged_data['Happiness Index']

# Perform Pearson correlation test
correlation_coefficient, p_value = stats.pearsonr(gdp, happiness)

print(f"Pearson correlation coefficient: {correlation_coefficient}")
print(f"P-value: {p_value}")

alpha = 0.05  # Significance level

if p_value < alpha:
    print("Reject the null hypothesis. There is a statistically significant correlation between GDP per capita and Happiness Index.")
else:
    print("Fail to reject the null hypothesis. There is no statistically significant correlation between GDP per capita and Happiness Index.")
```
