# Multi-variate analysis

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr


# Calculate the Pearson correlation coefficient and p-value
correlation_coefficient, p_value = pearsonr(df['Open_GOLD'], df['Close_GOLD'])

print(f"Pearson Correlation Coefficient: {correlation_coefficient}")
print(f"P-value: {p_value}")


# Interpret the results
alpha = 0.05  # Significance level

if p_value < alpha:
    print("Reject the null hypothesis. There is a statistically significant correlation between the opening and closing prices.")
    if correlation_coefficient > 0:
        print("The correlation is positive, indicating that higher opening prices tend to be associated with higher closing prices.")
    else:
        print("The correlation is negative, indicating that higher opening prices tend to be associated with lower closing prices.")

else:
    print("Fail to reject the null hypothesis. There is no statistically significant correlation between the opening and closing prices.")
```

# Model - 1 - 80/20 split - order 1: 2/1/2


```python
# importation of libraries
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
from math import sqrt
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error, r2_score
from math import sqrt
```

```python
train_size = int(0.8 * len(df))
train_data = df[:train_size]
test_data = df[train_size:]

# Build and train the ARIMA model
order = (2, 1, 2) # pdq
model = ARIMA(train_data['Close_GOLD'], order = order)
fit_model = model.fit()

# Predictions
predictions = fit_model.forecast(steps = len(test_data))

# model evaluation
rmse = sqrt(mean_squared_error(test_data['Close_GOLD'], predictions))

# Compute Metrics
# Compute Metrics
rmse = sqrt(mean_squared_error(test_data['Close_GOLD'], predictions)) # Changed 'test' to 'test_data'
mse = mean_squared_error(test_data['Close_GOLD'], predictions)       # Changed 'test' to 'test_data'
r2 = r2_score(test_data['Close_GOLD'], predictions)                 # Changed 'test' to 'test_data'
mean_residuals = (test_data['Close_GOLD'] - predictions).mean()      # Calculate residuals using test_data

print(f'Root Mean Squared Error (RMSE): {rmse}')
print(f"R-squared (R2): {r2}")
print(f"Mean of Residuals: {mean_residuals}")
print(f"Mean Squared Error (MSE): {mse}")

# prediction visalization
plt.figure(figsize=(12, 6))
# Convert 'Date_' to datetime if it's not already
train_data['Date_'] = pd.to_datetime(train_data['Date_'])
test_data['Date_'] = pd.to_datetime(test_data['Date_'])

# Use the 'x' and 'y' arguments directly instead of 'df'
sns.lineplot(x=train_data['Date_'], y=train_data['Close_GOLD'], label='Train')
sns.lineplot(x=test_data['Date_'], y=test_data['Close_GOLD'], label='Test')
sns.lineplot(x=test_data['Date_'], y=predictions, label='Predictions')
plt.title('Arima Model Prediction testing')
plt.show()
```

- 80/20 split.
- order:2,1,2

MSE: 0.9394 - Low MSE value.

RMSE: 0.939 - Low RMSE value.

Residuals: 0.882 - It suggests that the model is normally distributed.

# SARIMA modelling

1. Model 1 - 6 months, 80/20 split
order: 2/1/2

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error, r2_score
from math import sqrt
import matplotlib.pyplot as plt

# Define seasonality period (s=12 for monthly data), 12 time periods = 1 year
seasonal_period = 6

# Train-test split
train_size = int(len(df) * 0.8)
train, test = df[:train_size], df[train_size:]

# Fit SARIMA model (p, d, q)  (P, D, Q, s)
sarima_model = SARIMAX(
    train['Close_GOLD'],
    order=(2, 1, 2),  # ARIMA(p, d, q) parameters
    seasonal_order=(1, 1, 1, seasonal_period)  # Seasonal parameters (P, D, Q, s)
)
sarima_fit = sarima_model.fit()

# Make predictions
predictions = sarima_fit.forecast(steps=len(test))

# Calculate Residuals
residuals = test['Close_GOLD'] - predictions

# Compute Metrics
rmse = sqrt(mean_squared_error(test['Close_GOLD'], predictions))
mse = mean_squared_error(test['Close_GOLD'], predictions)
r2 = r2_score(test['Close_GOLD'], predictions)
mean_residuals = residuals.mean()

# Print Metrics
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2): {r2}")
print(f"Mean of Residuals: {mean_residuals}")

# Visualize predictions
plt.figure(figsize=(12, 6))
plt.plot(train.index, train['Close_GOLD'], label="Train Data")
plt.plot(test.index, test['Close_GOLD'], label="Test Data", color='orange')
plt.plot(test.index, predictions, label="SARIMA Predictions", linestyle="dashed", color='red')
plt.title('SARIMA Model Predictions')
plt.legend()
plt.show()
```

Root Mean Squared Error (RMSE): 0.9609881908895965 - Far from 0

Mean Squared Error (MSE): 0.9234983030292596 - FAr from 0

R-squared (R2): -0.06607311822339579 - Based off this result obtained with this model, we can safely say that this model is overfitting. Because of its negative value, it suggests that this model is not reliable and should not be used to predict the change of price.

Mean of Residuals: 0.17906029549620078 - Normally distributed

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error, r2_score
from math import sqrt
import matplotlib.pyplot as plt

# Define seasonality period (s=12 for monthly data), 12 time periods = 1 year
seasonal_period = 6

# Train-test split
train_size = int(len(df) * 0.8)
train, test = df[:train_size], df[train_size:]

# Fit SARIMA model (p, d, q)  (P, D, Q, s)
sarima_model = SARIMAX(
    train['Close_ETH-USD'],
    order=(5, 3, 1),  # ARIMA(p, d, q) parameters
    seasonal_order=(1, 1, 1, seasonal_period)  # Seasonal parameters (P, D, Q, s)
)
sarima_fit = sarima_model.fit()

# Make predictions
predictions = sarima_fit.forecast(steps=len(test))

# Calculate Residuals
residuals = test['Close_ETH-USD'] - predictions

# Compute Metrics
rmse = sqrt(mean_squared_error(test['Close_ETH-USD'], predictions))
mse = mean_squared_error(test['Close_ETH-USD'], predictions)
r2 = r2_score(test['Close_ETH-USD'], predictions)
mean_residuals = residuals.mean()

# Print Metrics
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2): {r2}")
print(f"Mean of Residuals: {mean_residuals}")

# Visualize predictions
plt.figure(figsize=(12, 6))
plt.plot(train.index, train['Close_ETH-USD'], label="Train Data")
plt.plot(test.index, test['Close_ETH-USD'], label="Test Data", color='orange')
plt.plot(test.index, predictions, label="SARIMA Predictions", linestyle="dashed", color='red')
plt.title('SARIMA Model Predictions')
plt.legend()
plt.show()
```

1. Model 1 - Period: 6 months, 80/20 split/ order: 5,5,3

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error, r2_score
from math import sqrt
import matplotlib.pyplot as plt

# Define seasonality period (s=12 for monthly data), 12 time periods = 1 year
seasonal_period = 6

# Train-test split
train_size = int(len(df) * 0.8)
train, test = df[:train_size], df[train_size:]

# Fit SARIMA model (p, d, q)  (P, D, Q, s)
sarima_model = SARIMAX(
    train['Close_BTC-USD'],
    order=(5, 5, 3),  # ARIMA(p, d, q) parameters
    seasonal_order=(1, 1, 1, seasonal_period)  # Seasonal parameters (P, D, Q, s)
)
sarima_fit = sarima_model.fit()

# Make predictions
predictions = sarima_fit.forecast(steps=len(test))

# Calculate Residuals
residuals = test['Close_BTC-USD'] - predictions

# Compute Metrics
rmse = sqrt(mean_squared_error(test['Close_BTC-USD'], predictions))
mse = mean_squared_error(test['Close_BTC-USD'], predictions)
r2 = r2_score(test['Close_BTC-USD'], predictions)
mean_residuals = residuals.mean()

# Print Metrics
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2): {r2}")
print(f"Mean of Residuals: {mean_residuals}")

# Visualize predictions
plt.figure(figsize=(12, 6))
plt.plot(train.index, train['Close_BTC-USD'], label="Train Data")
plt.plot(test.index, test['Close_BTC-USD'], label="Test Data", color='orange')
plt.plot(test.index, predictions, label="SARIMA Predictions", linestyle="dashed", color='red')
plt.title('SARIMA Model Predictions')
plt.legend()
plt.show()
```
